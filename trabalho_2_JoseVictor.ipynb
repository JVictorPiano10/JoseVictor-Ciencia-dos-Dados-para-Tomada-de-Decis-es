No histograma anterior, a barra que estava sobre os vaçlores próximo ou iguais a zero era muito maior que as outras barras combinadas, o histograma visualizou a frequência, e a máscara booleana a quantificou, a barra gigantesca estava próxima ao eixo y.

Podemos entender que essa alta quantidade de pagamentos zero é uma informação bastante importante, pois mostra que muitos clientes estão apenas pagando o mínimo, ou nem isso em vários meses.

```python
import numpy as np
import matplotlib.pyplot as plt
# Certifique-se de que a variável 'tabela' está carregada na memória

# Lista das colunas de valor pago (PAY_AMT)
colunas_valor_pago = [
    'PAY_AMT1', 'PAY_AMT2', 'PAY_AMT3',
    'PAY_AMT4', 'PAY_AMT5', 'PAY_AMT6'
]

# 1. Configuração do Plot: Cria um layout de 2 linhas e 3 colunas (2x3)
fig, ax_array = plt.subplots(2, 3, figsize=(18, 10))
eixos = ax_array.flatten()

# 2. Loop para aplicar a transformação e plotar
for i, coluna in enumerate(colunas_valor_pago):

    # a. FILTRA: Seleciona apenas os pagamentos diferentes de zero (maior que 0)
    # Reutilizamos o conceito da máscara booleana
    dados_coluna = tabela[coluna]
    dados_nao_zero = dados_coluna[dados_coluna > 0]

    # b. TRANSFORMAÇÃO: Aplica np.log10.
    # Embora np.log10(series) funcione diretamente (vetorização),
    # o conceito é o mesmo que usar .apply(np.log10) em uma Série
    dados_log = np.log10(dados_nao_zero)

    # c. PLOT: Cria o histograma
    eixos[i].hist(dados_log, bins=20, edgecolor='black', color='skyblue')

    # Rotação para rótulos legíveis (aplicando ao eixo, como corrigimos antes)
    eixos[i].tick_params(axis='x', rotation=45)

    eixos[i].set_title(f'Log10 de {coluna} (Pagamentos > 0)', fontsize=14)
    eixos[i].set_xlabel('Log10(Valor Pago)')
    eixos[i].set_ylabel('Frequência')

# 3. Exibição
plt.suptitle('Histogramas de Transformações Logarítmicas dos Pagamentos Diferentes de Zero', fontsize=16, y=1.02)
plt.tight_layout()
plt.show()
```

- Distribuição quase normal: Os dados não disitribuídos de forma assimétrica, mas são parecidfos com uma curva de sino (distribuição normal)

- Facilidade de Análise: A tranformação espreme os valores muito grandes e expande os valores pequenos, assim tornando os dados variados muiuto mais gerenciados

- Maior Concentração: Podemos notar que a maioria dos pagamentos diferentes de zero se concentra em torno de um ou dois picos centrais no gráfico transformado.
